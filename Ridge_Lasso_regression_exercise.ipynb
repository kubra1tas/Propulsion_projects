## Ridge Lasso Regression (Part 2)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
from sklearn import metrics
from scipy import stats
```

```python
plt.rcParams["figure.figsize"] = (8, 8) #Change matplotlib Box Size
plt.rcParams["font.size"] = 14 #Change matplotlib Font Size
```

```python
#To Do: load the dataset and remove column: quality
wine_data = pd.read_csv('/Users/iMotz/Propulsion/Machine_learning/data_sets/winequality-white.csv', sep = ';')
wine_data.head(5)
```

```python
wine_data_rem = wine_data.drop(columns = ['quality'])
```

```python
# To Do: Spit the data to X and y
X = wine_data_rem.drop(columns = ['alcohol'])
y = wine_data_rem['alcohol']
```

#### ACHTUNG! mistake in zscoring!!!

```python
# split data set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33, random_state=0)

# Normalise Data, it is better to normalize after split!

# standard scaler: StandardScaler()
# use .fit_transform(TRAIN)
# only .transform(TEST)


X_train = X_train.apply(stats.zscore)
X_test = X_test.apply(stats.zscore)

print('Training set has %d samples.' % len(X_train))
print('Testing set has %d samples.' % len(X_test))
```

```python
X_train
```

## Ridge Regression

```python
# Fit a ridge regression on the training data
# Use this model to predict the test data

#Ridge(alpha=1.0, fit_intercept=True, normalize=False, copy_X=True, max_iter=None, 
#      tol=0.001, solver=’auto’, random_state=None)

# fit_intercept = True, always better to use True! its how the input is related to output, worst case it is 0, doesnt matter

# alpha = 1 # Larger values specify stronger regularization. 

ridge_classifier = Ridge(alpha = 1, fit_intercept = True)

ridge_classifier.fit(X_train, y_train)

y_predict = ridge_classifier.predict(X_test)
y_predict
#pd.DataFrame(ridge.coef_,data.columns.values[0:10],columns=['Coefficients'])
```

```python
#Calculate the metrics
print('MAE',metrics.mean_absolute_error(y_test,y_predict))
print('MSE',metrics.mean_squared_error(y_test,y_predict))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
print('Explained Var Score',metrics.explained_variance_score(y_test,y_predict))
```
