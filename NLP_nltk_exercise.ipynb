```python
#filtered_sentence = [w for w in word_tokens if not w in stop_words] 
filt_sentence = []  
 
for i in range(0, len(raw_words)):
    curr_word_list = raw_words[i]
    # create list of filtered words
    filt_words = []
    for word in curr_word_list:
        if word not in stop_words_E: 
            filt_words.append(word) 
               
    # write into list           
    filt_sentence.append(filt_words)
```

## Create dictionary

```python
# get unique words with IDs
my_id2word = corpora.Dictionary(lemmat_sentence)
my_id2word[0]
```

```python
# Create Corpus
my_corpus_dict = [my_id2word.doc2bow(text) for text in lemmat_sentence]

print(my_corpus_dict[:1])
```

```python
 # Human readable format of corpus (term-frequency)
[[(my_id2word[id], freq) for id, freq in cp] for cp in my_corpus_dict[:1]]
```

## Build model and tune parameters

```python
# lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
#                                            id2word=id2word,
#                                            num_topics=20, 
#                                            random_state=100,
#                                            update_every=1,
#                                            chunksize=100,
#                                            passes=10,
#                                            alpha='auto',
#                                            per_word_topics=True)
```

```python
my_num_topics = 20

lda_model = gensim.models.ldamodel.LdaModel(corpus = my_corpus_dict, # counts of words per doc
                                           id2word = my_id2word, # list of unique words
                                           num_topics = my_num_topics, # number of clusters
                                           update_every = 1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
from pprint import pprint

pprint(lda_model.print_topics(num_words = 1))
```

### Evaluate model

```python
# Perplexity: how good is the model? The lower the better.
print('Perplexity: %.4f' % lda_model.log_perplexity(my_corpus_dict))
```

```python
# Compute Coherence Score
coherencemodel = gensim.models.CoherenceModel(model = lda_model, texts = lemmat_sentence, 
                                              dictionary = my_id2word, coherence='c_v')

coherence_lda = coherencemodel.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

```python
# get words to remove:
how_many_top = 100

top_freq_words = []

for i in range(0, my_num_topics):
    for j in range(0, how_many_top):
        
        top_freq_words.append(lda_model.show_topic(i, topn = how_many_top)[j][0])
    
top_freq_words
```
