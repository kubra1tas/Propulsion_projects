```python
import pandas as pd
import numpy as np
import scipy
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import cm

sns.set() # sets standard seaborn theme as default
```

```python
data = pd.read_csv('/Users/iMotz/Propulsion/PYTHON/Visualization/data/snb-data-zimoma-en-all-20170502_1430.csv')
# very useful date time conversion
data['Date'] = pd.to_datetime(data['Date'])
data.set_index(['Date','D0'], inplace=True)
#data.sort_index(inplace=True)
data = data.unstack()['Value']
data.head()
```

```python
# crop data
my_data = data['1TGT']
#my_data = pd.DataFrame(data = my_data, columns = )
```

```python
my_arrow = dict(arrowstyle = '->', color='black',lw = 2) 

# line chart of 1TGT series
plt.rcParams['figure.figsize'] = (10, 6)

fig, ax = plt.subplots()
ax.plot(data['1TGT'])

ax.axvspan('1991-01-01','1993-12-01' , facecolor='0.5', alpha=0.5, label = 'Recession in CH (1991-1993)')
ax.axvspan('2009-01-01', '2009-12-01', facecolor='0.5', alpha=0.3, label = 'Recession in CH (2009)')

# ax.axvline(x = '1973-11-01', linestyle = '--', c = 'black', label = 'Oil price shock (1973/11)')
# ax.axvline(x = '2008-09-01', linestyle = '--', c = 'orange', label = 'Lehman-Brothers collapse (2008/09)')
# ax.axvline(x = '2011-08-01', linestyle = '-', c = 'green', label = 'CHF/EUR floor intro (2011/08)')
# ax.axvline(x = '2015-01-01', linestyle = '-', c = 'grey', label = 'CHF/EUR floor removal (2015/01)')

ax.annotate('Oil price shock (1973/11)', xy = ('1973-11-01', 8), xytext = ('1973-11-01', 10.5),
            arrowprops = my_arrow, size = 12 )

ax.annotate('Lehman-Brothers\ncollapse (2008/09)', xy = ('2008-09-01', 3), xytext = ('2004-05-01', 10),
            arrowprops = my_arrow, size = 12 )

ax.annotate('CHF/EUR\nfloor intro\n(2011/08)', xy = ('2011-08-01', 0.5), xytext = ('2010-01-01', 7),
            arrowprops = my_arrow, size = 12 )

ax.annotate('CHF/EUR\nfloor removal\n(2015/01)', xy = ('2015-01-01', 0.3), xytext = ('2012-11-01', 3),
            arrowprops = my_arrow, size = 12 )

             
ax.axhline(y = 0, linestyle = ':', c = 'red',)
ax.legend(loc= 'upper right')
ax.set_xlabel('Time', size = 14)
ax.set_ylabel('Rate', size = 14)
None

#plt.savefig('/Users/iMotz/Propulsion/PYTHON/Visualization/Money_lines_4.png')
```
